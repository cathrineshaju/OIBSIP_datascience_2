```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x='Month',
    y='Unemployment Rate',
    hue='State',
    marker='o',
    data=df
)

plt.title("Monthly Unemployment Analysis")

plt.show()
```

```python
state_avg = df.groupby('State')['Unemployment Rate'].mean()

plt.figure(figsize=(6,6))

plt.pie(
    state_avg,
    labels=state_avg.index,
    autopct='%1.1f%%'
)

plt.title("Average Unemployment Distribution")
```
